```python
puzzle_input = read_input_numbers('day10.ipynb')
g2 = graph_joltages(puzzle_input)
nx.draw_networkx(g2, arrows=True, with_labels=True)
```

```python
matrix_power(nx.to_scipy_sparse_matrix(g2).todense(), 97)[-1, 0]
```

```python
g2.nodes()
```

```python
longest_path = nx.dag_longest_path(g2)
adjacency = nx.adjacency_matrix(g2).todense()
paths = 0
for path_length in range(len(longest_path) + 1):
    paths_of_length = matrix_power(adjacency, path_length)[-1, 0]
    print(f"Found {paths_of_length} paths of length {path_length}")
    paths += paths_of_length
print(f"{paths} total paths")
```
